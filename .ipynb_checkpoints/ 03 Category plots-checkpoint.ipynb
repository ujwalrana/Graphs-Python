```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Categorical Plots

```python
taxis = sns.load_dataset('taxis')
tips = sns.load_dataset('tips')
```

#### Bar

```python
sns.barplot(x='sex',y='tip', data= tips)
plt.show()
```

```python
sns.barplot(x='sex',y='tip', data= tips , estimator= sum)
```

```python
sns.barplot(x='sex',y='tip', data= tips , estimator= sum, hue='smoker')
```

```python
plt.figure(figsize=[3,3])
sns.barplot(x='sex',y='tip', data= tips , estimator= sum, hue='smoker')
```

```python
tips
```

```python
sns.barplot(x='sex',y='tip', data= tips , estimator= sum, hue='smoker', legend='auto')
plt.show()
```

Countplot

```python
plt.figure(figsize = [3,3])
sns.countplot(x='sex', data= tips , hue='smoker', legend='auto')
```

#### Box plot

```python
plt.figure(figsize= [3,3])
sns.boxplot(x = 'sex', y ='total_bill', data=tips)
plt.show()
```

```python
plt.figure(figsize= [3,3])
sns.boxplot(x = 'sex', y ='total_bill', data=tips, hue='smoker')
plt.show()
```

#### Histogram

```python
sns.histplot(x='sex', data=tips)
```

```python
sns.histplot(x='day', data=tips)
```

```python
sns.histplot(x='total_bill', data=tips)
```

#### Violin

```python
sns.violinplot(data= tips, x='day', y='total_bill')  # boxPlot or histogram ek sath milte hea violin graph meh
```

#### Strip 

```python
sns.stripplot(data= tips, x='day', y='total_bill') # indiviual points/values marks yah show hote hea stripplot meh 
```

#### Swarm

```python
sns.swarmplot(data= tips, x='day', y='total_bill') # dono sectiom ko combine krne ki try ki gyi hea swarm plot meh 
plt.show() 
```

#### Distribution Plots

```python
sns.histplot(data=tips, x='tip') # histogram is used to count data
plt.show()
```

```python
plt.figure(figsize=[3,3])
sns.histplot(data=tips, x='total_bill', kde=True)
plt.show()
```

#### Kernal Density Estimator plot

```python
plt.figure(figsize=[3,3])
sns.kdeplot(data=tips, x='total_bill', hue='sex')
plt.show()
```

#### Distribution Plots

```python
plt.figure(figsize=[3,3]) # this function not use in future so we do this with histogram without any error
sns.distplot(x=tips['total_bill'])
plt.show()
```

#### Matrix plot

```python
taxis = sns.load_dataset('taxis')
```

```python
taxis.corr(numeric_only=True) # in this we found the correlation 
```

```python
plt.figure(figsize=[3,3])
sns.heatmap(taxis.corr(numeric_only=True))
```

```python
plt.figure(figsize=[3,3])
sns.heatmap(taxis.corr(numeric_only=True), cmap='viridis')
```

#### Cluster Map

```python
plt.figure(figsize=[3,3])
sns.clustermap(taxis.corr(numeric_only=True))
```
