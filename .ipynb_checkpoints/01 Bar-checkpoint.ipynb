```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'Category': "A B C D".split(),
        'values' : [10, 20, 15, 8]
       }
```

#### Column Bar Graph

```python
sns.set_theme(style="whitegrid", palette="pastel")

sns.barplot(data=data, x='Category', y='values')
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.barplot(data=tips, x='day', y='total_bill', estimator='mean')
```

#### Row Bar Graph

```python
taxis = sns.load_dataset('taxis')
```

```python
plt.figure(figsize=(10, 30))  # Adjust the height as needed

sns.barplot(x='total', y = 'dropoff_zone', data=taxis, estimator=sum)
plt.tight_layout()
```

```python
sns.barplot(data=tips, y='sex', x='total_bill', estimator=sum)
```
